# Data Cleaning

```python
plt.figure(figsize=(12,7))
sns.boxplot(x='Pclass' ,y='Age' , data=data,palette='winter')
```

```python
def cal_age(cols):
    Age=cols[0]
    Pclass= cols[1]
    
    if pd.isnull(Age):
        if Pclass ==1:
            return 37
        elif Pclass ==2:
            return 29
        else:
            return 24
    else:
        return Age
    
            
```

```python
data['Age']=data[['Age','Pclass']].apply(cal_age , axis=1)
```
